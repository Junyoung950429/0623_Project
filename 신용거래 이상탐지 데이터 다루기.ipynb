# 구매 금액 z-score

```python
# cc_num 데이터 개수 조회
cc_df['cc_num'].value_counts()
```

```python
# cc_num 값마다 amt의 평균, 표준편차 저장
amt_info = cc_df.groupby('cc_num')['amt'].agg(['mean','std']).reset_index()
```

```python
amt_info
```

```python
# Q. cc_num 컬럼을 기준으로, cc_df와 amt_info 데이터를 합쳐서 cc_df에 저장합니다.
cc_df = cc_df.merge(amt_info, how='left', on='cc_num')
```

```python
# Q. 결제금액(amt)의 z-score를 계산하여 amt_z 컬럼에 저장합니다.
cc_df['amt_z']  = (cc_df['amt'] - cc_df['amt'].mean()) / cc_df['amt'].std()
```

```python
# 사기 거래 데이터(is_fraud=1) 조회
cc_df[cc_df['is_fraud'] == 1]
```

```python
# z-score 계산이 완료되었으니 mean, std 컬럼 제거
cc_df.drop(['mean','std'], axis = 1, inplace = True)
```

```python
# cc_num, category 컬럼의 값마다 결제금액(amt)의 평균, 표준편차 cat_info에 저장
cat_info = cc_df.groupby(['cc_num','category'])['amt'].agg(['mean','std']).reset_index()
```

```python
cat_info.head()
```

```python
# Q. cc_num 컬럼을 기준으로, cc_df와 cat_info 데이터를 합쳐서 cc_df에 저장합니다.
cc_df = cc_df.merge(cat_info, how='left', on=['cc_num', 'category'])
```

```python
# Q. 결제금액(amt)의 z-score를 계산하여 cat_amt_z 컬럼에 저장
cc_df['cat_amt_z']  = (cc_df['amt'] - cc_df['amt'].mean()) / cc_df['amt'].std()
```

```python
# z-score 계산이 완료되었으니 mean, std 컬럼 제거
cc_df.drop(['mean','std'], axis =1 , inplace = True)
```

# 시간 분석

```python
# Q. trans_date_trans_time 컬럼에서 시간 값을 추출하여 hour 컬럼에 저장합니다.
# datetime으로 변환 후, 시간추출한 값을 hour 열로 추가
cc_df['hour'] = pd.to_datetime(cc_df['trans_date_trans_time']).dt.hour
```

```python
cc_df.head()
```

```python
cc_df.tail()
```

```python
# 결제시간을 morning, afternoon, night, evening으로 분류하기 위해 함수를 정의합니다.

def hour_func(x):
    if (x >= 6) & (x < 12):
        return 'morning'
    elif (x >= 12) & (x < 18):
        return 'afternoon'
    elif (x >= 18) & (x < 23):
        return 'night'
    else:
        return 'evening'
```

```python
# Q. hour 컬럼에 hour_func 함수를 적용한 값을 hour_cat 컬럼에 저장

cc_df['hour_cat'] = cc_df['hour'].apply(hour_func)
```

```python
cc_df['hour_cat'].unique()
```

```python
# Q. 시간대별로 데이터 수를 계산하여 출력합니다.

cc_df['hour_cat'].value_counts()
```

```python
# cc_num 컬럼의 값마다 amt 컬럼의 데이터 수를 계산하여 all_cnt 변수로 저장

all_cnt = cc_df.groupby('cc_num')['amt'].count().reset_index()
```

```python
# cc_num, hour_cat 컬럼의 값마다 amt 컬럼의 데이터 수를 계산하여 hour_cnt 변수로 저장

hour_cnt = cc_df.groupby(['cc_num','hour_cat'])['amt'].count().reset_index()
```

```python
all_cnt.head()
```

```python
hour_cnt.head()
```

```python
# Q. cc_num 컬럼을 기준으로 hour_cnt와 all_cnt를 합칩니다.
# (hour_cnt에서 left merge를 수행합니다.)

hour_cnt = hour_cnt.merge(all_cnt, how='left', on='cc_num')
```

```python
hour_cnt.head()
```

```python
# Q. amt_x와 amt_y 컬럼의 이름을 각각 hour_cnt, total_cnt로 변경합니다.

hour_cnt = hour_cnt.rename({'amt_x': 'hour_cnt', 'amt_y': 'total_cnt'}, axis =1)
```

```python
hour_cnt.head()
```

```python
# Q. 전체 거래 건수(total_cnt)에 대한 시간대별 거래 건수(hour_cnt)의 비율을 계산하여 hour_perc 컬럼에 저장합니다.

hour_cnt['hour_perc'] = hour_cnt['hour_cnt'] / hour_cnt['total_cnt']
```

```python
hour_cnt.head(10)
```

```python
hour_cnt.tail(10)
```

```python
cc_df.head()
```

```python
# hour_cnt의 카드넘버, 시간대, 시간대별 거래건수 비율만 가져온다
hour_cnt = hour_cnt[['cc_num','hour_cat','hour_perc']]
```

```python
# Q. cc_num, hour_cat 컬럼을 기준으로 cc_df와 hour_cnt 데이터를 합칩니다.

cc_df = cc_df.merge(hour_cnt, how='left', on=['cc_num', 'hour_cat'])
```

```python
cc_df.head()
```

```python
# 시간 관련 feature 분석이 완료되었으므로 불필요해진 컬럼들을 제거합니다.

cc_df.drop(['trans_date_trans_time', 'hour', 'hour_cat'], axis =1 , inplace = True)
```
